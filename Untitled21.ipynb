```python
# ============================================================
# üöÄ UNSW-NB15 Dataset - Multiclass Attack Classification (Optimized)
# ============================================================

# Step 1: Install Libraries
!pip install xgboost imbalanced-learn lightgbm catboost scikit-learn pandas matplotlib seaborn --quiet

# Step 2: Import Required Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

# ============================================================
# Step 3: Load Dataset
# ============================================================

# Upload manually in Colab before running this (UNSW_NB15_training-set.csv and testing-set.csv)
# from google.colab import files
# uploaded = files.upload()

train_data = pd.read_csv("UNSW_NB15_training-set.csv")
test_data = pd.read_csv("UNSW_NB15_testing-set.csv")

print("‚úÖ Data Loaded Successfully!")
print("Train shape:", train_data.shape)
print("Test shape:", test_data.shape)

# ============================================================
# Step 4: Combine for Preprocessing
# ============================================================

data = pd.concat([train_data, test_data])

# Drop irrelevant ID column if present
if 'id' in data.columns:
    data = data.drop(columns=['id'])

# Encode categorical columns
cat_cols = data.select_dtypes(include=['object']).columns
for col in cat_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))

# Split back into train/test
train = data.iloc[:len(train_data), :]
test = data.iloc[len(train_data):, :]

# ============================================================
# Step 5: Define Features and Target (attack_cat)
# ============================================================

X_train = train.drop(columns=['attack_cat'])
y_train = train['attack_cat']
X_test = test.drop(columns=['attack_cat'])
y_test = test['attack_cat']

# Encode target labels
le_attack = LabelEncoder()
y_train = le_attack.fit_transform(y_train)
y_test = le_attack.transform(y_test)

# ============================================================
# Step 6: Normalize Features
# ============================================================

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ============================================================
# Step 7: Handle Class Imbalance using SMOTE
# ============================================================

print("\n‚öñÔ∏è Before SMOTE:", np.bincount(y_train))
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
print("‚öñÔ∏è After SMOTE:", np.bincount(y_train_bal))

# ============================================================
# Step 8: Train Optimized XGBoost Model
# ============================================================

xgb_opt = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=10,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=0.1,
    reg_lambda=1,
    random_state=42,
    eval_metric='mlogloss'
)

print("\nüöÄ Training XGBoost Model (Optimized)...")
xgb_opt.fit(X_train_bal, y_train_bal)

# ============================================================
# Step 9: Evaluate Model
# ============================================================

y_pred_xgb = xgb_opt.predict(X_test)
acc = accuracy_score(y_test, y_pred_xgb)

print(f"\n‚úÖ XGBoost Model Accuracy: {acc * 100:.2f}%")
print("\nüìä Classification Report:")
class_names = [str(c) for c in le_attack.classes_]

# Print classification report
print(classification_report(y_test, y_pred_xgb, target_names=class_names))

# ============================================================
# Step 10: Confusion Matrix Visualization
# ============================================================

cm = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=False, cmap='Blues')
plt.title("XGBoost - Confusion Matrix (Multiclass)")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# ============================================================
# Step 11: Feature Importance Plot
# ============================================================

importances = xgb_opt.feature_importances_
indices = np.argsort(importances)[-10:]

plt.figure(figsize=(7,5))
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [train.drop(columns=['attack_cat']).columns[i] for i in indices])
plt.title("üî• Top 10 Important Features (XGBoost)")
plt.xlabel("Feature Importance Score")
plt.ylabel("Feature")
plt.show()
```
